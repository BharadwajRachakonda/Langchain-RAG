```python
from langchain_openai import OpenAI
from google.colab import userdata


openai_api_key = userdata.get('OPENAI_API_KEY')
llm = OpenAI(api_key=openai_api_key,model='gpt-3.5-turbo-instruct') #max_completion_tokens, temperature
result = llm.invoke("Hello")
print(result.content)
```

#**Open Source Models**
> *Hugging Face Inference API, Running Locally*

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from google.colab import userdata


huggingface_api_key = userdata.get('HUGGINGFACEHUB_ACCESS_TOKEN')
llm = HuggingFaceEndpoint(huggingfacehub_api_token = huggingface_api_key, repo_id="deepseek-ai/DeepSeek-V4-Pro", task="text-generation")
model = ChatHuggingFace(llm = llm)

result = model.invoke("Hello?")
print(result.content)
```

```python
# Download and use locally the model is large so I didn't download
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline
from google.colab import userdata


huggingface_api_key = userdata.get('HUGGINGFACEHUB_ACCESS_TOKEN')
llm = HuggingFacePipeline.from_model_id(model_id="deepseek-ai/DeepSeek-V4-Pro", task="text-generation", pipeline_kwargs={
    "temperature": 0.5,
    "max_new_tokens": 512
})
model = ChatHuggingFace(llm = llm)

result = model.invoke("wait a min, who are you!?")
print(result.content)
```

#**Embedding Models**

```python
#for single text
from langchain_openai import OpenAIEmbeddings
from google.colab import userdata


openai_api_key = userdata.get('OPENAI_API_KEY')
embedding = OpenAIEmbeddings(api_key = openai_api_key, model='text-embedding-3-large', dimensions=32)

result = embedding.embed_query("Delhi is the capital of India")
print(str(result))
```

```python
#for entire doc
from langchain_openai import OpenAIEmbeddings
from google.colab import userdata


openai_api_key = userdata.get('OPENAI_API_KEY')
embedding = OpenAIEmbeddings(api_key = openai_api_key, model='text-embedding-3-large', dimensions=32)
documents = [
    "hello",
    "hi",
    "bye"
]
result = embedding.embed_documents(documents)
print(str(result))
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from google.colab import userdata
#downloading model and run

huggingface_api_key = userdata.get('HUGGINGFACEHUB_ACCESS_TOKEN')
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
text = "Yo this is embedding"

result = embedding.embed_query(text) #embedding.embed_documents(documents)
print(str(result))
```

#Eample Usage
###*Find Simillar document*

```python
from sklearn.metrics.pairwise import cosine_similarity
from langchain_huggingface import HuggingFaceEndpointEmbeddings
import numpy as np
from google.colab import userdata


huggingface_api_key = userdata.get("HUGGINGFACEHUB_ACCESS_TOKEN")
embedding = HuggingFaceEndpointEmbeddings(huggingfacehub_api_token = huggingface_api_key,model = "sentence-transformers/all-MiniLM-L6-v2")
documents = [
    "India capital is delhi",
    "AP capital is Amaravati",
    "Telangana capital is Hyderabad"
]
text = "what is the capital of India"
doc_embeddings = embedding.embed_documents(documents)
text_embedding = embedding.embed_query(text)
similaritys = cosine_similarity([text_embedding],doc_embeddings)[0]
best = max(list(enumerate(similaritys)), key=lambda x: x[1])
best_ele = documents[best[0]]
print(f"{best_ele=}")
```
