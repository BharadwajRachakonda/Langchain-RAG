Template

```python
from langchain_core.prompts import PromptTemplate

#template
template = PromptTemplate(
    """
    My name is {name} student of {college}
    """,
    input_variables=["name", "college"]
)
```

```python
prompt = template.invoke({
    "name": "variable_1",
    "college": "variable_2"
}, validate_template = True) # you can also use f strings but using template you can ensure that all variables are provided

#model.invoke(prompt)
```

Template for chat models

```python
#prompt template for chat models
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
'''bellow wont work'''
# template = ChatPromptTemplate([
#     SystemMessage(content="You are a nice AI bot acting as a {role}."),
#     HumanMessage(content="What is the meaning of {topic}?"),
# ])
'''this works'''
template = ChatPromptTemplate([
    ('system', "You are a nice AI bot acting as a {role}."),
    ('human', "What is the meaning of {topic}?"),
])

prompt = template.invoke({
    "role": "psycho",
    "topic": "life"
})

print(prompt)
```

Feeding history of other chats

```python
# to open and feed previous chat history
from langchain_core.prompts import MessagesPlaceholder, ChatPromptTemplate

template = ChatPromptTemplate([
    ('system', "You are a nice AI bot acting as a {role}."),
    MessagesPlaceholder(variable_name="chat_history"),
    ('human', "{input}"),
])

chat_history = []
with open("chat_history.txt") as f:
  i = f.readlines()
  chat_history.append(i)

prompt = template.invoke({
    "role": "psycho",
    "chat_history": chat_history,
    "input": "I didn't get you"
})

print(prompt)
```
