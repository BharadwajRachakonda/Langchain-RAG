###TextLoader
> *every document loader converts connestopding information into **Document object***

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('hello.txt', encoding='utf8') #encoding is optional
data = loader.load() # we get a list of document objects (for all loaders)
print(data)
```

###PyPDFLoader
> *if pdf has 25 pages resulting list will have 25 documents*

| Use Case | Recommended Loader |
|-----------|-------------------|
| Simple, clean PDFs | `PyPDFLoader` |
| PDFs with tables/columns | `PDFPlumberLoader` |
| Scanned/Image PDFs | `UnstructuredPDFLoader` or `AmazonTextractPDFLoader` |
| Need layout and image data | `PyMuPDFLoader` |
| Want best structure extraction | `UnstructuredPDFLoader` |

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader('hello.pdf') #only works with text pdf
data = loader.load() # we get a list of document objects (for all loaders)
print(data)
```

###DirectoryLoader
> *load multiple documents from a directory*

| Glob Pattern | What It Loads |
|-------------|---------------|
| `**/*.txt` | All `.txt` files in all subfolders |
| `*.pdf` | All `.pdf` files in the root directory |
| `data/*.csv` | All `.csv` files in the `data/` folder |
| `**/*` | All files (any type, all folders) |

```python
from langchain_community.document_loaders import  DirectoryLoader, PyPDFLoader

loader = DirectoryLoader(
    path = 'books', # more like ./books in node.js
    glob = '*.pdf',
    loader_cls = PyPDFLoader
)

docs = loader.load()
```

###Lazy loading
> *If there are multiple files loading takes time, so this returns a generator so it loads on demand*

```python
from langchain_community.document_loaders import  DirectoryLoader, PyPDFLoader

loader = DirectoryLoader(
    path = 'books', # more like ./books in node.js
    glob = '*.pdf',
    loader_cls = PyPDFLoader
)

docs = loader.lazy_load()
```

###WebBaseLoader
> *load and extract info from webpages, URLs for javascript heavy pages use SeleniumURLLoader*

```python
# for single url you get only one document, you can also pass list of urls then you get list of documents
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader('https://www.google.com')
data = loader.load()
print(data)
```

###CSVLoader
> *you get a document for each row*

```python
# for single url you get only one document, you can also pass list of urls then you get list of documents
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader('hello.csv')
data = loader.load()
print(data)

#you can find other document loaders in there website, you can also make a custome data loaders read documentation
```
