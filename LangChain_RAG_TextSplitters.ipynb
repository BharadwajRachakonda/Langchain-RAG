#**Text Splitters**
> *These are used to split text into smaller chunks, so that LLM can handle them effectively, this is because some LLMs and Embedding Models have limit on the amount of input size, splitting helps process documents that would otherwise exceide these limits, and it is also better memory efficient.*

> *Length Based, Text Structure Based, Document Structure Based, Semantic Meaning Based*

| Task | Why Splitting Helps |
|------|----------------------|
| Embedding | Short chunks yield more accurate vectors |
| Semantic Search | Search results point to focused info, not noise |
| Summarization | Prevents hallucination and topic drift |

###Length Based

```python
from langchain_text_splitters import CharacterTextSplitter

text = "Helo this is bulma, how are you"
splitter = CharacterTextSplitter(chunk_size=5, chunk_overlap=1, separator='')
splitter.split_text(text)
```

###Text Structure Based

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text = "Helo this is bulma, how are you"
splitter = RecursiveCharacterTextSplitter(chunk_size=10, chunk_overlap=0)
splitter.split_text(text)
```

###Document Structure Based

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter, Language

text = "..Some Python Code here"
splitter = RecursiveCharacterTextSplitter.from_language(language=Language.PYTHON, chunk_size=170, chunk_overlap=0) # used for codes, markdown.....
splitter.split_text(text)
```
